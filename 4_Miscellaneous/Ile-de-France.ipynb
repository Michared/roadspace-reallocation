# Ile-de-France

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import osmnx as ox
```

```python
colors = ['#00A6D6', '#4DD7FF', '#C4F2FF']
line_color = '#017188'
```

```python
P = ox.geocode_to_gdf('Paris')
PC = pd.concat([ox.geocode_to_gdf(dep) for dep in ['val-de-marne', 'hauts-de-seine', 'seine-saint-denis']]) 
GC = pd.concat([ox.geocode_to_gdf(dep) for dep in ['seine-et-marne', 'essonne', 'yvelines', "val d'oise"]]) 
```

```python
for zone, color in zip([P, PC, GC], colors):
    zone['color'] = color
```

```python
IDF = pd.concat([P, PC, GC])
bounds = IDF.copy()
bounds['geometry'] = bounds['geometry'].apply(lambda x: x.boundary)
bounds['color'] = line_color
```

```python
to_plot = pd.concat([IDF, bounds])
fig = to_plot.plot(color=to_plot['color'], linewidth=0.2)
plt.axis('off')
```

```python
plt.gcf().savefig(r'./results'+'/IDF.png', dpi=800, bbox_inches='tight')
```
